Let's see we can open pulse data of the 1724 and 1730 and see if there is an offset/sync

```python
import strax
import amstrax
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import numba
import os
```

```python
mongo_url = f'mongodb://{os.environ["MONGO_USER"]}:{os.environ["MONGO_PASSWORD"]}@127.0.0.1:27017/admin'
output_folder = '/data/xenon/xams/run11/processed/'
def new_context():
    """Create strax context that can access the runs db"""
    # We use exactly the logic of straxen to access the runs DB;
    # this avoids duplication, and ensures strax can access the runs DB if we can
    st = strax.Context(
        storage=amstrax.RunDB(
            mongo_url=mongo_url,
            mongo_dbname='run',
            mongo_collname='runs_new',
            runid_field='number',
            new_data_path=output_folder),
        config=amstrax.contexts.common_config,
        **amstrax.contexts.common_opts)
    return st
st = new_context()
```

```python
st.context_config['forbid_creation_of'] = amstrax.DAQReader.provides
```

```python
rr = st.get_array(run_id = '000089', targets='raw_records', seconds_range = (0,1))
rr_sipm = st.get_array(run_id = '000089', targets='raw_records_sipm', seconds_range = (0,1))
```

let's just plot some pulses (this is very easy since there is just one channel, otherwise one would need to do a pproper job by looking at `strax.record_links`

```python
for pi in range(40):
    for r in rr[7*pi:7*(pi+1)]:
        plt.plot(r['time']+r['dt']*np.arange(r['length']), r['data'][:r['length']])
    for r in rr_sipm[2*pi:2*(pi+1)]:
        plt.plot(r['time']+r['dt']*np.arange(r['length']), r['data'][:r['length']])
    plt.title(rr[7*pi]['time']-rr_sipm[2*pi]['time'])
    plt.grid()
    plt.show()
```

```python
links = np.array(strax.record_links(rr))
for rec_i, r in enumerate(rr):
    start, stop = links[:,rec_i]
    plt.plot(r['time']+r['dt']*np.arange(r['length']), r['data'][:r['length']])
    if stop == -1:
        plt.show()
    if rec_i > 100:
        break
```
